# 해제 사유 발생날짜
해제 사유 발생 여부로 바꿀 수 있을 것 같음.
날짜를 알 수가 없으면 결측치 대신에 0,1 으로 바꾸면 될 것 같음


```python
train["해제사유발생유무"] = np.where(train['해제사유발생일'].notna(),1,0)
```

# Drop
해제 사유 발생일과 해제사유발생유무, 중개사소재지 Drop

```python
train['is_test'] = 0
test['is_test'] = 1
concat = pd.concat([train, test]) 
```

```python
concat.columns
```

```python
concat.drop(["해제사유발생유무", "해제사유발생일", "중개사소재지"], axis=1, inplace=True)
```

```python
concat.head()
```

```python
modified_train = concat[concat["is_test"]==0]
modified_test = concat[concat["is_test"]==1]
```

```python
modified_train.head()
modified_train.drop(["is_test"], axis=1, inplace=True)
```

```python
modified_test.head()
modified_test.drop(["is_test"], axis=1, inplace=True)
```
